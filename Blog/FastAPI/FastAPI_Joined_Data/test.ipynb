```python
from sqlalchemy import create_engine
import pandas as pd
import json
```

```python
engine = create_engine("mysql+pymysql://nook:2909@192.168.10.22/test")
```

```python
sql = """
        SELECT
            a.id AS id,
            a.customer AS customer,
            a.date_start AS `datetime`,
            e.id AS engineer_id,
            e.engineer_name AS engineer_name 
        FROM
            addjob a
            LEFT JOIN engineer AS e ON a.engineer = e.engineer_name
    """
# Assuming you have a connect() function to establish the database connection

# Read data from SQL query
df = pd.read_sql(sql, engine)

# Group data by engineer_id and engineer_name
grouped = df.groupby(['engineer_id', 'engineer_name'])

# Create a list to store grouped data
grouped_data = []

# Loop through groups and create a dictionary for each engineer
for (engineer_id, engineer_name), group_data in grouped:
    engineer_dict = {
        "engineer_id": engineer_id,
        "engineer_name": engineer_name,
        "jobs": group_data.to_dict(orient='records')
    }
    grouped_data.append(engineer_dict)

# Print or return the final grouped_data
grouped_data
```
